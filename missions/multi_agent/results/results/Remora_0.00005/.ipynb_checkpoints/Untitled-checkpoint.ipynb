```python
import pandas as pd
```

```python
df = pd.read_csv("grace_goals.csv")
df['Time'] = 600 - df['Time']
```

```python
result_df = None

def fun(x, interval=25, deadline=600):
    
    
    result = {
             "index": [],
             "Time":[],
             "Goals(#)":[]}
    
    no_of_goals = 0
    for i in range(0,deadline,interval):
        rows = x[ (x["Time"] >= i) & (x["Time"] < (i+ interval))]
        if len(rows) > 0:
            no_of_goals = rows['Goals(#)'].sum()

        result["Time"].append(i+ interval)
        result["Goals(#)"].append(no_of_goals)
        
        if result_df:
            result_df
        
    return pd.Series(result)
        
    
    
df.groupby("index")[['Time', 'Goals(#)', "index"]].apply(lambda x : fun(x))
```
